```python
#1.1 Imprima una lista de todos los numeros capicuas de 1 a 9, cuantos hay?
num=['1','2','3','4','5','6','7','8','9']


palindromes = [''.join(i) for i in itertools.product(num,repeat=9) if i == i[::-1]]
print palindromes
a=len(palindromes)
print "Existen",a, "numeros palindromos en este rango"

```

```python
#1.2 Imprima una lista de todos los capic ́as de 9 dígitos que tambien son capicuas en rep. binariarepresentaci ́n binaria. Cada l ́
binarios=[]
#Lista de numeros binarios
for item in palindromes:
    item=int(item)
    item=np.binary_repr(item)    
    binarios.append(item)
    
print binarios


    
```

```python
#Retorna una lista con los numeros binarios que son capicuas
x=[]
for i in binarios:
    if i[::-1] == i:
        x.append(i)
print x
```

```python
#1.3 Verifique si una cadena de texto es palíndroma
Texto=input("Digite el texto entre comillas:") #Insertar texto en comillas
uTexto = unicode(Texto, 'utf8') #Convierte en unicode el input del usuario
sinacc=unicodedata.normalize('NFKD', uTexto).encode('ascii', 'ignore') #elimina acentos
crc_sp=re.sub(r"\W", "", sinacc) #Remueve \t y \n
crc_sp=''.join( c for c in sinacc if  c in 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz') #Crea una cadena de solo caracteres
crc_sp=crc_sp.lower() #Transforma todo a lower case
check=[crc_sp]

#Recorre la cadena de caracteres y verifica si el texto es un palíndromo
for i in check:
    reverso=i[::-1]
    palindrom=(reverso==i)
    if palindrom:
        print "El texto es un palindromo"
    else:
        print "El texto no es un palindromo"

```
